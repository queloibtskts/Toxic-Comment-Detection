```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv(r'../data/train.csv')
```

```python
train.describe()
```

```python
# histogram
plt.hist(train['target'], edgecolor = 'black', bins = 10)

# Add labels
plt.title('Histogram of Toxicity Label', fontname='Times New Roman', fontsize = 15)
plt.xlabel('Toxicity', fontname='Times New Roman')
plt.ylabel('Frequency (million)', fontname='Times New Roman')
plt.show
```

```python
toxic = train[train.target >= 0.5].copy()
```

```python
toxic.shape
```

```python
# Pie chart
labels = 'Toxic', 'Non-toxic'
sizes = [(train.shape[0] - toxic.shape[0]), toxic.shape[0]]
#explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig, ax = plt.subplots()
ax.pie(sizes, explode=(0, 0.2), 
       labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
fig.savefig('labelDistribution.png')
```
